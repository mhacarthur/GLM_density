```python
import glob
import numpy as np
import pandas as pd

import cartopy
import cartopy.feature as cfeature
import cartopy.crs as ccrs

import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
raios_list = glob.glob('/mnt/Data/Data/GLM/SA/Model_Alan_Format/*.dat')
raios_list = sorted(raios_list)
raios_list[0:4]
```

```python
total_raios = np.zeros([len(raios_list)])
for t in range(len(raios_list)):
    data = pd.read_csv(raios_list[t],header=None,sep='\t')
    total_raios[t] = len(data)
```

```python
index = np.where(total_raios>=34000)[0]
print(index)
```

```python
raios1 = pd.read_csv(raios_list[index[0]],header=None,sep='\t')
raios2 = pd.read_csv(raios_list[index[1]],header=None,sep='\t')
raios3 = pd.read_csv(raios_list[index[2]],header=None,sep='\t')
raios4 = pd.read_csv(raios_list[index[3]],header=None,sep='\t')
```

```python
name_tmp = raios_list[index[0]].split('/')[-1]
year = name_tmp[5:9]
month = name_tmp[9:11]
day = name_tmp[11:13]
hour1 = name_tmp[13:15]
minute1 = name_tmp[15:17]
print(f'Time 1: {year} {month} {day} {hour1}:{minute1}')

name_tmp = raios_list[index[1]].split('/')[-1]
year = name_tmp[5:9]
month = name_tmp[9:11]
day = name_tmp[11:13]
hour2 = name_tmp[13:15]
minute2 = name_tmp[15:17]
print(f'Time 1: {year} {month} {day} {hour2}:{minute2}')

name_tmp = raios_list[index[2]].split('/')[-1]
year = name_tmp[5:9]
month = name_tmp[9:11]
day = name_tmp[11:13]
hour3 = name_tmp[13:15]
minute3 = name_tmp[15:17]
print(f'Time 1: {year} {month} {day} {hour3}:{minute3}')

name_tmp = raios_list[index[3]].split('/')[-1]
year = name_tmp[5:9]
month = name_tmp[9:11]
day = name_tmp[11:13]
hour4 = name_tmp[13:15]
minute4 = name_tmp[15:17]
print(f'Time 1: {year} {month} {day} {hour4}:{minute4}')
```

```python
title_size = 8
fontsize_axis = 8
s_raios = 0.2

proj = ccrs.PlateCarree()

fig = plt.figure(figsize=(6,6),dpi=200)
gs = gridspec.GridSpec(2,2)

ax1 = plt.subplot(gs[0, 0], projection = proj)
ax1.coastlines(linewidth=0.5,color='black',zorder=4)
ax1.add_feature(cfeature.COASTLINE,linewidth=0.5,color='black',zorder=4)
ax1.add_feature(cfeature.BORDERS,linewidth=0.5,color='black',zorder=4)
ax1.set_extent([-86,-30,-42,9])
ax1.scatter(raios1[9],raios1[8],s=s_raios,zorder=5)
gl = ax1.gridlines(crs=proj,draw_labels=True,linewidth=0.6,color='gray',alpha=0.6,linestyle='--')
gl.top_labels = False
gl.bottom_labels = False
gl.left_labels = True
gl.right_labels = False
gl.xlabel_style = {'size': fontsize_axis, 'color': 'black'}
gl.ylabel_style = {'size': fontsize_axis, 'color': 'black'}
# ax1.annotate(day+'/'+month + '  ' + hour1+':'+minute1, xy=(-48.5,-36), xytext=(-48.5,-36),fontsize=8)
ax1.annotate(day+'/'+month, xy=(-39,-34), xytext=(-39,-34),fontsize=8)
ax1.annotate(hour1+':'+minute1, xy=(-39,-38), xytext=(-39,-38),fontsize=8)

ax2 = plt.subplot(gs[0, 1], projection = proj)
ax2.coastlines(linewidth=0.5,color='black',zorder=4)
ax2.add_feature(cfeature.COASTLINE,linewidth=0.5,color='black',zorder=4)
ax2.add_feature(cfeature.BORDERS,linewidth=0.5,color='black',zorder=4)
ax2.set_extent([-86,-30,-42,9])
ax2.scatter(raios2[9],raios2[8],s=s_raios,zorder=5)
gl = ax2.gridlines(crs=proj,draw_labels=True,linewidth=0.6,color='gray',alpha=0.6,linestyle='--')
gl.top_labels = False
gl.bottom_labels = False
gl.left_labels = False
gl.right_labels = False
gl.xlabel_style = {'size': fontsize_axis, 'color': 'black'}
gl.ylabel_style = {'size': fontsize_axis, 'color': 'black'}
# ax2.annotate(day+'/'+month + '  ' + hour2+':'+minute2, xy=(-48.5,-36), xytext=(-48.5,-36),fontsize=8)
ax2.annotate(day+'/'+month, xy=(-39,-34), xytext=(-39,-34),fontsize=8)
ax2.annotate(hour2+':'+minute2, xy=(-39,-38), xytext=(-39,-38),fontsize=8)

ax3 = plt.subplot(gs[1, 0], projection = proj)
ax3.coastlines(linewidth=0.5,color='black',zorder=4)
ax3.add_feature(cfeature.COASTLINE,linewidth=0.5,color='black',zorder=4)
ax3.add_feature(cfeature.BORDERS,linewidth=0.5,color='black',zorder=4)
ax3.set_extent([-86,-30,-42,9])
ax3.scatter(raios3[9],raios3[8],s=s_raios,zorder=5)
gl = ax3.gridlines(crs=proj,draw_labels=True,linewidth=0.6,color='gray',alpha=0.6,linestyle='--')
gl.top_labels = False
gl.bottom_labels = True
gl.left_labels = True
gl.right_labels = False
gl.xlabel_style = {'size': fontsize_axis, 'color': 'black'}
gl.ylabel_style = {'size': fontsize_axis, 'color': 'black'}
# ax3.annotate(day+'/'+month + '  ' + hour3+':'+minute3, xy=(-48.5,-36), xytext=(-48.5,-36),fontsize=8)
ax3.annotate(day+'/'+month, xy=(-39,-34), xytext=(-39,-34),fontsize=8)
ax3.annotate(hour3+':'+minute3, xy=(-39,-38), xytext=(-39,-38),fontsize=8)

ax4 = plt.subplot(gs[1, 1], projection = proj)
ax4.coastlines(linewidth=0.5,color='black',zorder=4)
ax4.add_feature(cfeature.COASTLINE,linewidth=0.5,color='black',zorder=4)
ax4.add_feature(cfeature.BORDERS,linewidth=0.5,color='black',zorder=4)
ax4.set_extent([-86,-30,-42,9])
ax4.scatter(raios4[9],raios4[8],s=s_raios,zorder=5)
gl = ax4.gridlines(crs=proj,draw_labels=True,linewidth=0.6,color='gray',alpha=0.6,linestyle='--')
gl.top_labels = False
gl.bottom_labels = True
gl.left_labels = False
gl.right_labels = False
gl.xlabel_style = {'size': fontsize_axis, 'color': 'black'}
gl.ylabel_style = {'size': fontsize_axis, 'color': 'black'}
# ax4.annotate(day+'/'+month + '  ' + hour4+':'+minute4, xy=(-48.5,-36), xytext=(-48.5,-36),fontsize=8)
ax4.annotate(day+'/'+month, xy=(-39,-34), xytext=(-39,-34),fontsize=8)
ax4.annotate(hour4+':'+minute4, xy=(-39,-38), xytext=(-39,-38),fontsize=8)

plt.subplots_adjust(left=None, bottom=None, right=None, top=None, hspace=0.04, wspace=0.02)

salida = '../images/GLM_15min_01_2019_I.png'
print(salida)
plt.savefig(salida,transparent = False,bbox_inches ='tight',pad_inches = 0.01, facecolor='white')
```
